# Sistema de segundo orden

La función de transferencia de un sistema de segundo orden puede expresarse en la forma general:

H(s) = K / [(s^2) + (2ζωn s) + (ωn^2)]

Donde: <br />
K es la ganancia del sistema <br />
ζ es el factor de amortiguamiento (también conocido como coeficiente de amortiguamiento) <br />
ωn es la frecuencia natural del sistema <br />

La frecuencia natural del sistema se define como la frecuencia a la que el sistema oscilaría si no hubiera amortiguamiento y se calcula como:

ωn = sqrt(k/m)

Donde: <br />
k = es la constante de fuerza del sistema <br />
m = es la masa del sistema <br />
El factor de amortiguamiento ζ indica la cantidad de amortiguamiento presente en el sistema y puede variar de 0 (sin amortiguamiento) a 1 (amortiguamiento crítico). Cuando ζ es menor que 1, el sistema es subamortiguado, lo que significa que los valores de salida oscilan alrededor del valor de equilibrio antes de converger. Cuando ζ es igual a 1, el sistema está críticamente amortiguado y la respuesta se estabiliza lo más rápido posible sin oscilaciones. Cuando ζ es mayor que 1, el sistema está sobreamortiguado y la respuesta se estabiliza más lentamente y sin oscilaciones.

La función de transferencia del sistema de segundo orden se puede expresar en términos de su frecuencia natural y factor de amortiguamiento, lo que permite analizar su comportamiento dinámico y predecir su respuesta ante diferentes entradas.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

## Valores de la funcion de transferencia

```python
K = 1
zeta = 0.5
wn = 2*np.pi*10
num = [K*wn**2]
den = [1, 2*zeta*wn, wn**2]
sys = signal.TransferFunction(num, den)
```

## Señal de entrada

```python
t, y = signal.step(sys)
```

## Graficando

```python
fig = plt.figure(figsize=(10,10))
plt.plot(t, y)
plt.xlabel('Tiempo (s)')
plt.ylabel('Salida')
plt.title('Respuesta del sistema de segundo orden')
plt.grid()
plt.show()
```
